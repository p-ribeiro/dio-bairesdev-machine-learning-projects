```python
from PIL import Image
from IPython.display import display
```

```python
img = Image.open("lena.png")

print("Displaying the original image:")
display(img)
```

```python
# converting an image to shades of grey

gray_img = img.convert("L")

print("Displaying the image in shades of grey:")
display(gray_img)
```

```python
# We can apply a threshold to turn the image into binary black and white

threshold = 128 # half way between 0-255 (lower values means more white)
# applying the threshold so every values over 128 are pure black and below are pure white
bw_img = gray_img.point(lambda x: 255 if int(x) > threshold else 0, mode="1")

print("Displaying the image in binary black and white:")
display(bw_img)
```
